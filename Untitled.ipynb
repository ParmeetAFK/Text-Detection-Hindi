## Importing modules 

```python
# Library for OCR
import easyocr

# Computer Vision library for drawing Rectangles and writing text on Images
import cv2

# To display the Image
from matplotlib import pyplot as plt

# To naviagte the folders / Images
import os
```

## Path to folder full of Images

```python
# Describe the path for folder where all Images are stored

FOLDER_PATH = r"C:\Users\Parme\Documents\Untitled Folder\IMAGES"
```

## Create Instance of EASYOCR

```python
# Create Instance of EasyOCR Library
# Two Parameters
# 1st - Describe the list of codes for language you want to be detected
# 2nd - Boolean value if you want to use GPU for processing 

reader = easyocr.Reader(['hi'], gpu=False)
```

## Function for Text Detection

```python
# Function returns a list of detections words in an image
# 2 Parameters
# 1st = Path for Image to be scanned for words
# 2nd = EasyOCR instance



def find_words(img_path , reader):
    result = reader.readtext(img_path)
    return result
```

## Function for Display Detection

```python

# Font to write on Images
font = cv2.FONT_HERSHEY_SIMPLEX


# Draw Rectangle around the detected Text 
# 2 Paramters
# 1st = List of all detections
# 2nd = Image

def draw_detected(detection , img):
    spacer = 100

    # Extract Upper left and Bottom Right co-ordinates of detections
    top_left = tuple((int(detection[0][0][0]) , int(detection[0][0][1])))
    bottom_right = tuple((int(detection[0][2][0]) , int(detection[0][2][1])))
    
    # Extract Text from the detection 
    text = detection[1]
    
    
    # Draw Rectangle on Image
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    
    # Write the detected text on Image
    img = cv2.putText(img,text,(20,spacer), font, 0.5,(0,255,0),2,cv2.LINE_AA)
    spacer+=15
    
    
```

## Display Summary for Image

```python
# function to display summary of detections

def display_summary(detections):
    
    # Calculates total number of detections
    sum_of_detections = len(detections)
    
    words = []
    for i in detections:
        words.append(i[1])
    
    
    print("------------ SUMMARY -----------------")
    print(" Total Detections = " , sum_of_detections)
    print(" Detected Texts = " , words)
    print("--------------------------------------")
```

## Main 

```python



# Loop  through all images in folder
for img in os.listdir(FOLDER_PATH):
    
    
    # Creates Path for Image inside Folder
    img_path = FOLDER_PATH + "\ ".strip() + img
    
    # Opens Image using OpenCV
    img = cv2.imread(img_path)
    
    # All detections are stored in detections variable
    detections = find_words(img_path, reader)
    
    
    # Loop through detections and draw rectaangle around them
    for detection in detections:
        draw_detected(detection,img)
    
    
    # Display the Image
    plt.imshow(img)
    plt.show()
    
    # Display the summary
    display_summary(detections)
    

```
